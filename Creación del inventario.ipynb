# Creación del inventario


El presente notebook buscará diseñar un sistema de inventario simple, el cual deberá permitir sus modificaciones respectivas.

```python
inventario = {'lapices':[23, 300],
              'lapiceras': [15, 1000],
              'carpetas': [8, 3000],
              'tijeras': [17, 2000] }

while True:
    modo_uso = input("Desea consultar (c) o modificar(m) un producto?   ").lower()
    if modo_uso in ("c", "m"):
        if modo_uso == "c":
            print(f'Consulte el precio de {list(inventario.keys())}')
            producto = input("Escriba un producto:  ").lower()
            if producto in inventario:
                stock = inventario[producto][0]
                precio = inventario[producto][1]
                print(f'En este momento tiene {stock} {producto} a {precio} cada una')
            else:
                print(f'el producto {producto} no se encuentra en el inventario')
        else:
            modificacion = True
            while modificacion:
                 print(f'Los productos a modificar son {list(inventario.keys())}')
                 producto = input("Escriba un producto:  ").lower()
                 if producto in inventario:
                     print(f'Ha elegido el producto {producto}')
                     try:
                         cambio_stock = int(input("Escriba el cambio de stock:  "))
                         inventario [producto] [0] += cambio_stock
                         cambio_precio = float(input("Escriba el cambio del precio:  "))
                         inventario [producto] [1] += cambio_precio
                         stock = inventario[producto][0]
                         precio = inventario[producto][1]
                         print(f'Ahora tiene {stock} {producto} a {precio} cada una')
                         modificacion = False
                     except:
                         print('No ha escrito un número válido')
                 else:
                     print(f'el producto {producto} no se encuentra en el inventario')               
    else:
        if modo_uso == "exit":
            break
        else:
            print('respuesta no válida, responda con c o m')
            print('responda "exit" para salir')
```
