# Paso de temperaturas

La presente función buscará actuar de convertor entre grados celsius y grados farenheit utilizando la fórmula **F = C x 1.8 + 32**

El proceso constará de 3 partes:
- una entrada utilizando la función input( ) para pedir el número que se desea transformar en farenheit.
- la primer parte del proceso será transformar en float utilizando la función float( ). A ese valor se le aplicará la fórmula, dando la transformación como resultado.
- la salida será un print( ) que indicará a qué valor de farenheit pertenece.

```python
entrada = input("Ingrese los grados celsius que quiere transformar en farenheit: ")
num_float = float(entrada)
resultado = num_float * 1.8 + 32
print(f'el valor indicado es de {resultado} celsius')
```
