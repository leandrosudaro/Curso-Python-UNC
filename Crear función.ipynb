# Crear función

La presente actividad buscará crear una función modular utilizando el argumento *args. La función tendrá el objetivo de sumar todos los argumentos presentados, valiéndose para ello de un bucle for que sumara los argumentos de forma seriada; para luego dividir el resultado por la cantidad de argumentos.


```python
def promedio_ponderado(a, b, *args):
    a = a + b
    cantidad = 2
    for num in args:
        a = a + num
        cantidad += 1
    resultado = a/cantidad
    return resultado

    
```
